```python
import pandas as pd

# Load dataset
df = pd.read_csv("nilai-gizi.csv")

# Kolom yang dibutuhkan
cols = [
    'name',
    'energy_kcal',
    'protein_g',
    'fat_g',
    'carbohydrate_g',
    'fiber_g',
    'sodium_mg'
]

df = df[cols].copy()

# -------------------------
# FILTER MENU JADI
# -------------------------

MASAKAN_KEYWORDS = [
    'masakan',
    'soto',
    'rawon',
    'gulai',
    'rendang',
    'semur',
    'tumis',
    'sayur',
    'opor',
    'sate',
    'pepes',
    'balado',
    'sup',
    'sop',
    'bakar',
    'goreng',
    'pecel',
    'gado',
    'urap',
    'capcay',
    'nasi',
    'padamaran',
    'kambose',
    'parede',
    'kaparende',
    'keumamah'
]

FORBIDDEN = [
    'mentah',
    'segar',
    'kering',
    'tepung',
    'daun',
    'buah',
    'susu',
    'milk',
    'tea',
    'coffee',
    'chatime',
    'coklat',
    'biskuit',
    'wafer',
    'keripik',
    'es krim',
    'ice blended',
    'saus',
    'kecap',
    'yogurt',
    'minuman'
]

# ubah ke lowercase
df['name_lower'] = df['name'].str.lower()

# hanya menu masakan
df_menu = df[
    df['name_lower'].apply(
        lambda x:
        any(k in x for k in MASAKAN_KEYWORDS)
        and
        not any(f in x for f in FORBIDDEN)
    )
].copy()

# bersihkan nama
df_menu['nama_masakan'] = (
    df_menu['name']
    .str.replace(', masakan', '', regex=False)
    .str.replace(', sayur', '', regex=False)
    .str.strip()
    .str.title()
)

# hapus duplikat
df_menu.drop_duplicates(
    subset=['nama_masakan'],
    inplace=True
)

# id
df_menu.insert(
    0,
    'id',
    range(1, len(df_menu)+1)
)

# placeholder bahan
df_menu['bahan'] = 'Belum tersedia'

# urutan kolom
hasil = df_menu[
    [
        'id',
        'nama_masakan',
        'bahan',
        'energy_kcal',
        'protein_g',
        'fat_g',
        'carbohydrate_g',
        'fiber_g',
        'sodium_mg'
    ]
]

hasil.to_csv(
    'menu_mbg.csv',
    index=False
)

print("Jumlah menu:", len(hasil))
print(hasil.head())
```

```python
import pandas as pd
import random

# ==========================
# LOAD DATA
# ==========================

karbo = pd.read_csv("karbo.csv")
protein = pd.read_csv("protein.csv")
sayur = pd.read_csv("sayur.csv")

# ==========================
# FILTER DATA KARBO
# ==========================

karbo = karbo[
    (karbo["energy_kcal"] >= 100) &
    (karbo["carbohydrate_g"] >= 20)
]

# ==========================
# FILTER DATA PROTEIN
# ==========================

protein = protein[
    (protein["protein_g"] >= 8)
]

# Hindari bahan mentah dan olahan tidak sehat
forbidden_protein = [
    "mentah",
    "asin",
    "kering",
    "kaleng",
    "sosis",
    "nugget",
    "abon"
]

protein = protein[
    ~protein["name"].str.lower().str.contains(
        "|".join(forbidden_protein),
        na=False
    )
]

# ==========================
# FILTER DATA SAYUR
# ==========================

sayur = sayur[
    (sayur["fiber_g"] >= 1)
]

forbidden_sayur = [
    "segar",
    "mentah",
    "tepung"
]

sayur = sayur[
    ~sayur["name"].str.lower().str.contains(
        "|".join(forbidden_sayur),
        na=False
    )
]

# ==========================
# GENERATE MENU
# ==========================

hasil = []

for i in range(3000):

    k = karbo.sample(1).iloc[0]
    p = protein.sample(1).iloc[0]
    s = sayur.sample(1).iloc[0]

    total_energy = (
        k["energy_kcal"] +
        p["energy_kcal"] +
        s["energy_kcal"]
    )

    total_protein = (
        k["protein_g"] +
        p["protein_g"] +
        s["protein_g"]
    )

    total_carb = (
        k["carbohydrate_g"] +
        p["carbohydrate_g"] +
        s["carbohydrate_g"]
    )

    total_fat = (
        k["fat_g"] +
        p["fat_g"] +
        s["fat_g"]
    )

    total_fiber = (
        k["fiber_g"] +
        p["fiber_g"] +
        s["fiber_g"]
    )

    total_sodium = (
        k["sodium_mg"] +
        p["sodium_mg"] +
        s["sodium_mg"]
    )

    hasil.append({
        "menu_name": f"{k['name']} + {p['name']} + {s['name']}",

        "karbo": k["name"],
        "protein": p["name"],
        "sayur": s["name"],

        "bahan": f"{k['name']}, {p['name']}, {s['name']}",

        "energy_kcal": round(total_energy, 1),
        "protein_g": round(total_protein, 1),
        "carbohydrate_g": round(total_carb, 1),
        "fat_g": round(total_fat, 1),
        "fiber_g": round(total_fiber, 1),
        "sodium_mg": round(total_sodium, 1)
    })

# ==========================
# DATAFRAME
# ==========================

menu = pd.DataFrame(hasil)

# Hilangkan kombinasi duplikat
menu.drop_duplicates(
    subset=["karbo", "protein", "sayur"],
    inplace=True
)

# ==========================
# FILTER STANDAR MBG
# ==========================

menu = menu[
    (menu["energy_kcal"] >= 500) &
    (menu["energy_kcal"] <= 900) &
    (menu["protein_g"] >= 20) &
    (menu["carbohydrate_g"] >= 50) &
    (menu["fiber_g"] >= 5) &
    (menu["sodium_mg"] <= 800)
]

# ==========================
# SORT TERBAIK
# ==========================

menu["mbg_score"] = (
    menu["protein_g"] * 0.4 +
    menu["fiber_g"] * 0.3 +
    menu["energy_kcal"] / 100 * 0.2 -
    menu["sodium_mg"] / 1000 * 0.1
)

menu = menu.sort_values(
    by="mbg_score",
    ascending=False
)

# ==========================
# SIMPAN
# ==========================

menu.to_csv(
    "menu_harian_mbg.csv",
    index=False
)

print("Jumlah menu valid:", len(menu))
print(menu.head())
```

```python
import pandas as pd

karbo = pd.read_csv("karbo_bersih.csv")
protein = pd.read_csv("protein_bersih.csv")
sayur = pd.read_csv("sayur_bersih.csv")

hasil = []

for _, k in karbo.iterrows():
    for _, p in protein.iterrows():
        for _, s in sayur.iterrows():

            total_energy = (
                k["energy_kcal"] +
                p["energy_kcal"] +
                s["energy_kcal"]
            )

            total_protein = (
                k["protein_g"] +
                p["protein_g"] +
                s["protein_g"]
            )

            total_carb = (
                k["carbohydrate_g"] +
                p["carbohydrate_g"] +
                s["carbohydrate_g"]
            )

            total_fat = (
                k["fat_g"] +
                p["fat_g"] +
                s["fat_g"]
            )

            total_fiber = (
                k["fiber_g"] +
                p["fiber_g"] +
                s["fiber_g"]
            )

            total_sodium = (
                k["sodium_mg"] +
                p["sodium_mg"] +
                s["sodium_mg"]
            )

            # filter gizi masuk akal untuk MBG
            if (
                450 <= total_energy <= 900 and
                total_protein >= 15
            ):

                hasil.append({
                    "menu_name":
                        f"{k['name']} + {p['name']} + {s['name']}",

                    "karbo": k["name"],
                    "protein": p["name"],
                    "sayur": s["name"],

                    "bahan":
                        f"{k['name']}, {p['name']}, {s['name']}",

                    "energy_kcal":
                        round(total_energy, 1),

                    "protein_g":
                        round(total_protein, 1),

                    "carbohydrate_g":
                        round(total_carb, 1),

                    "fat_g":
                        round(total_fat, 1),

                    "fiber_g":
                        round(total_fiber, 1),

                    "sodium_mg":
                        round(total_sodium, 1)
                })

menu = pd.DataFrame(hasil)

menu.drop_duplicates(
    subset=["karbo", "protein", "sayur"],
    inplace=True
)

menu.to_csv(
    "menu_harian_mbg.csv",
    index=False
)

print("Jumlah menu:", len(menu))
print(menu.head())
```

```python
import pandas as pd

df = pd.read_csv("menu_mbg_level2.csv")

# =========================
# WHITELIST KARBO
# =========================
karbo_valid = [
    "nasi",
    "lontong",
    "jagung",
    "ubi",
    "singkong",
    "ketela"
]

# =========================
# WHITELIST PROTEIN
# =========================
protein_valid = [
    "ayam",
    "telur",
    "ikan",
    "tempe",
    "tahu",
    "sapi",
    "bakso"
]

# =========================
# WHITELIST SAYUR
# =========================
sayur_valid = [
    "bayam",
    "kangkung",
    "wortel",
    "brokoli",
    "cap cai",
    "capcay",
    "gado",
    "karedok",
    "sayur lebui",
    "anyang",
    "ares"
]

# =========================
# FILTER WHITELIST
# =========================

df = df[
    df["karbo"].str.lower().str.contains(
        "|".join(karbo_valid),
        na=False
    )
]

df = df[
    df["protein"].str.lower().str.contains(
        "|".join(protein_valid),
        na=False
    )
]

df = df[
    df["sayur"].str.lower().str.contains(
        "|".join(sayur_valid),
        na=False
    )
]

# =========================
# BLACKLIST PRODUK OLAHAN
# =========================

blacklist = [

    "indomie",
    "sarimi",
    "supermi",
    "mie abc",
    "mie sedaap",
    "pop mie",
    "lemonilo",

    "royco",
    "khong guan",
    "malkist",

    "chatime",
    "ultra milk",
    "ultrajaya",
    "you c100",

    "kopi",
    "coffee",
    "teh",

    "es krim",
    "banana split",
    "yogurt",

    "saus",
    "kecap",
    "sambal",

    "subway",
    "yoshinoya",
    "church texas",

    "ice blended",
    "mooncake",
    "silverqueen",
]

for kata in blacklist:

    df = df[
        ~df["menu_name"]
        .str.lower()
        .str.contains(kata, na=False)
    ]

# =========================
# HAPUS MENU EKSTREM
# =========================

df = df[
    (df["energy_kcal"] >= 400)
]

df = df[
    (df["energy_kcal"] <= 700)
]

df = df[
    (df["protein_g"] >= 15)
]

df = df[
    (df["fiber_g"] >= 1)
]

df = df[
    (df["sodium_mg"] <= 800)
]

df = df[
    (df["fat_g"] <= 25)
]

# =========================
# HAPUS DUPLIKAT
# =========================

df = df.drop_duplicates(
    subset=["karbo", "protein", "sayur"]
)

# =========================
# SORT
# =========================

df = df.sort_values(
    by=[
        "protein_g",
        "fiber_g"
    ],
    ascending=False
)

# =========================
# SIMPAN
# =========================

df.to_csv(
    "menu_mbg_level3.csv",
    index=False
)

print("Jumlah menu akhir :", len(df))
print(df.head(20))
```
